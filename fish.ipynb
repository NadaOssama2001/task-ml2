```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Fish.csv")
df.head()
```

```python
df.info()
```

```python
X=df.iloc[:,-1].values.reshape(-1,1)
y=df.iloc[:,1].values.reshape(-1,1)
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_model= PolynomialFeatures(degree=5)
poly_model.fit_transform(X)
```

```python
plt.scatter(X,y ,color='red')
plt.plot(X,model.predict(X),color='blue')
plt.title('linear')
plt.xlabel('width')
plt.ylabel('weight')
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_model= PolynomialFeatures(degree=5)
X_poly=poly_model.fit_transform(X)
line_poly=LinearRegression()
line_poly.fit(X_poly,y)
```

```python
plt.scatter(X,y ,color='red')
plt.plot(X,line_poly.predict(poly_model.fit_transform(X)),color='blue')
plt.title('linear')
plt.xlabel('width')
plt.ylabel('weight')
plt.show()
```
